```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
# Load the Excel file
# Load the data sheet into a DataFrame
df = pd.read_excel("C:/Users/ravil/Jupyter note book/assigment/Clustering/EastWestAirlines.xlsx")
```

```python
# Visualizing with boxplots to identify outliers
plt.figure(figsize=(15, 10))
df_numeric = df.drop(columns=['ID#'])  # Exclude ID column for plotting
sns.boxplot(data=df_numeric)
plt.title('Boxplot for Numeric Features to Detect Outliers')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scaling the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_numeric)
```

```python
# Convert scaled features back to a DataFrame
df_scaled = pd.DataFrame(scaled_features, columns=df_numeric.columns)
```
